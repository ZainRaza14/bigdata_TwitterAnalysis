### Positive and Negative Words Separation

```python
import nltk

#nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

```

```python
sid.polarity_scores('now')
```

```python
#check_words = ['happy','sad','ok','now']
```

```python
pos_words = []

neg_words = []

tmp = []

for word in c_Words:
    pol_word = sid.polarity_scores(word)
    if pol_word['pos'] == 1:
        pos_words.append(word)
    elif pol_word['neg'] == 1:
        neg_words.append(word)
    else:
        del tmp[:]
        tmp.append(pol_word['pos'])
        tmp.append(pol_word['neg'])
        tmp.append(pol_word['neu'])
        max_tmp_ind = tmp.index(max(tmp))
        if max_tmp_ind == 0:
            pos_words.append(word)
        else:
            neg_words.append(word)
        
```

### Writing pos and neg to files

```python
with open('pos.txt', 'w') as filehandle:
    for word in pos_words:
        filehandle.write('%s\n' % word)
```

```python
with open('neg.txt', 'w') as filehandle:
    for word in neg_words:
        filehandle.write('%s\n' % word)
```
