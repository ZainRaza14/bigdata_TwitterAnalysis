### Positive and Negative Words Separation

```python
sid.polarity_scores('now')
```

```python
#check_words = ['happy','sad','ok','now']
```

```python
pos_words = []

neg_words = []

tmp = []

for word in c_Words:
    pol_word = sid.polarity_scores(word)
    if pol_word['pos'] == 1:
        pos_words.append(word)
    elif pol_word['neg'] == 1:
        neg_words.append(word)
    else:
        del tmp[:]
        tmp.append(pol_word['pos'])
        tmp.append(pol_word['neg'])
        tmp.append(pol_word['neu'])
        max_tmp_ind = tmp.index(max(tmp))
        if max_tmp_ind == 0:
            pos_words.append(word)
        else:
            neg_words.append(word)
        
```

### Creating new files

```python


tweets = []

tweets_text = []

for n in range(29, 59):
    file = "../../Twitter Data/01/00/" + str(n) + ".json"
    for line in open(file, 'r'):
        tweets.append(json.loads(line))


```

```python
for i in range(len(tweets)):
    if 'text' in tweets[i]:
        tweets_text.append(tweets[i]['text'])
    
c_Words = cleanWords(tweets_text)
```

```python
print(len(c_Words))
```

```python
pos_words = []

neg_words = []

tmp = []

for word in c_Words:
    pol_word = sid.polarity_scores(word)
    if pol_word['pos'] == 1:
        pos_words.append(word)
    elif pol_word['neg'] == 1:
        neg_words.append(word)
    else:
        del tmp[:]
        tmp.append(pol_word['pos'])
        tmp.append(pol_word['neg'])
        tmp.append(pol_word['neu'])
        max_tmp_ind = tmp.index(max(tmp))
        if max_tmp_ind == 0:
            pos_words.append(word)
        else:
            neg_words.append(word)
        
```

```python
finalNegWords = []

for neg_word in neg_words:
    if neg_word in words.words():
        finalNegWords.append(neg_word)
        
```
