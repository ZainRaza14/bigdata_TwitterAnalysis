```python

import os
import json 

directory = '../../Twitter Data/rest of data'
folders = [f for f in os.listdir(directory) if "." not in f]
folders.sort()



def filter_data(file_):
    data = []
    for line in open(file_, 'r'):
        data.append(json.loads(line))

    raw_tweets = []
    for i in range(len(data)):
        if 'lang' in data[i] and 'text' in data[i]:
            if data[i]['lang'] == 'en':
                raw_tweets.append(data[i])
    return raw_tweets


tweets = []
for folder in folders:
    data_files = [f for f in os.listdir(directory+'/'+folder) if ".bz2" not in f and '.DS' not in f]
    data_files.sort()
    for file_ in data_files:
        file_path = directory + '/' + folder + '/' + file_
        print(file_path)
        tweets.extend(filter_data(file_path))

```

```python
pos_text = []

neg_text = []

neu_text = []

tmp = []

for tweet_text in finalTweets:
    pol_text = sid.polarity_scores(tweet_text)
    if pol_text['pos'] == 1:
        pos_text.append(tweet_text)
    elif pol_text['neg'] == 1:
        neg_text.append(tweet_text)
    elif pol_text['neu'] == 1:
        neu_text.append(tweet_text)
    else:
        del tmp[:]
        tmp.append(pol_text['pos'])
        tmp.append(pol_text['neg'])
        tmp.append(pol_text['neu'])
        max_tmp_ind = tmp.index(max(tmp))
        if max_tmp_ind == 0:
            pos_text.append(tweet_text)
        elif max_tmp_ind == 1:
            neg_text.append(tweet_text)
        else:
            neu_text.append(tweet_text)
        
```

```python
import time
```

```python

start_time = time.time()

count_vect = CountVectorizer()

X_train_counts = count_vect.fit_transform(train['text'])

tfidf_transformer = TfidfTransformer()

X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python

start_time = time.time()


clf = MultinomialNB().fit(X_train_tfidf, train['label'])

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
X_new_counts = count_vect.transform(test['text'])

X_new_tfidf = tfidf_transformer.transform(X_new_counts)

```

```python
predicted = clf.predict(X_new_tfidf)
```

```python
np.mean(predicted == test['label'])
```

### LR

```python
start_time = time.time()

lr_text = LogisticRegression(verbose=1, solver='liblinear', penalty='l2')

lr_text.fit(X_train_tfidf,train['label'])

print("--- %s seconds ---" % (time.time() - start_time))

predicted_lr = lr_text.predict(X_new_tfidf)

np.mean(predicted_lr == test['label'])
```

### SVM

```python
start_time = time.time()

svm_text = LinearSVC()

svm_text.fit(X_train_tfidf, train['label'])

print("--- %s seconds ---" % (time.time() - start_time))

predicted_svm = svm_text.predict(X_new_tfidf)

np.mean(predicted_svm == test['label'])
```
