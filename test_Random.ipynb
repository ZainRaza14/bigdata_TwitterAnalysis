```python
import pandas as pd

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import GaussianNB

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.naive_bayes import MultinomialNB

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier

from sklearn.svm import LinearSVC
```

```python
pos = []

neg = []

f = open("pos.txt", "r")
for x in f:
    pos.append(x)
```

```python
print(pos)
```

```python
f.close()
```

```python
neg_f = open("neg.txt", "r")
for y in neg_f:
    neg.append(y)
```

```python
print(len(neg))
```

```python
print(len(pos))
```

```python
check_Data = pd.DataFrame()
```

```python
check_Data['words'] = pos
```

```python
check_Data1 = pd.DataFrame()
```

```python
check_Data1['words'] = neg
```

```python
for x in range(0, 15717):
    check_Data['label'] = 1
```

```python
for x in range(0, 59484):
    check_Data1['label'] = 0
```

```python
tmp_fram = [check_Data, check_Data1]
```

```python
finalDF = pd.concat(tmp_fram)
```

```python
finalDF
```

```python
finalDF.sample(frac=1)
```

```python
train, test = train_test_split(finalDF, test_size=0.25)
```

```python
print(train['label'])
```

```python
count_vect = CountVectorizer()

X_train_counts = count_vect.fit_transform(train['words'])
```

```python
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)

X_train_tf = tf_transformer.transform(X_train_counts)
```

```python
tfidf_transformer = TfidfTransformer()

X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

### Naive Bayes

```python


clf = MultinomialNB().fit(X_train_tfidf, train['label'])

```

```python
X_new_counts = count_vect.transform(test['words'])

X_new_tfidf = tfidf_transformer.transform(X_new_counts)

```

```python
predicted = clf.predict(X_new_tfidf)
```

```python
np.mean(predicted == test['label'])
```

### Logistic Regression

```python
lr_text = LogisticRegression(verbose=1, solver='liblinear',random_state=0, C=5, penalty='l2',max_iter=1000)
```

```python
lr_text.fit(X_train_tfidf,train['label'])
```

```python
predicted_lr = lr_text.predict(X_new_tfidf)

np.mean(predicted_lr == test['label'])
```

### SVM

```python
svm_text = LinearSVC()
```

```python
svm_text.fit(X_train_tfidf, train['label'])
```

```python
predicted_svm = svm_text.predict(X_new_tfidf)

np.mean(predicted_svm == test['label'])
```
